```python
import pandas as pd
data_input = pd.read_csv('tmp/validation_real_data.csv')
data_input.info()
```

```python
from pycaret.regression import load_model, predict_model
loaded_model_simple = load_model('models/model_wo_ohe_2024-12-03')
```

```python
# Realiza la predicción
predicciones = predict_model(loaded_model_simple, data=data_input)
predicciones
```

```python
predicciones.to_csv('tmp/predictions.csv', index=False)
```
